## Code Cell 1 (5%) :  Library and data import. Show the summary of the input data.

```python
import pandas as pd
import spacy
import nltk
from spacy import displacy
from collections import Counter
nlp = spacy.load("en_core_web_lg")
df = pd.read_csv('restaurant_reviews_az.csv')
df.info()
```

* Imported all the libraries
* Loaded the trained model
* Read the dataset to df

## Code Cell 2 (5%) –Select the 1 star reviews and 5 star reviews from the dataset. 

```python
df_selected = df[(df['stars']==1) | (df['stars']==5)]
df_selected.shape
```

```python
df_selected.head(5)
```

```python
# df_selected = df_selected.loc[:1500, ['text','stars']]
# df_selected
```

* There are total 32933 rows after filtering out the reviews with 1 and 5 stars

## Code Cell 3 (30%) - Apply necessary text processing techniques on the selected reviews. 

### Tokenization

```python
df_selected['nlp_doc'] = df_selected['text'].apply(nlp)
df_selected
```

* Applied the pre trained model to text column

#### Sentence and Word Tokenization

```python
df_selected['sentence'] = df_selected['nlp_doc'].apply(lambda x: [sent for sent in nlp(x).sents])
df_selected['word'] = df_selected['nlp_doc'].apply(lambda x: [token for token in nlp(x)])
df_selected
```

* Performed Sentence Tokentisation
* Performed Word Tokenization

#### Stop Words Removal

```python
df_selected['cleaned_words'] = df_selected['word'].apply(lambda tokens: [token for token in tokens if not token.is_stop])
df_selected
```

* Removed the stop words

### Part-Of-Speech Tagging

```python
df_selected['pos'] = df_selected['nlp_doc'].apply(lambda doc: [(token.text, token.pos_) for token in doc])
df_selected
```

* Performed POS tagging and saved it pos column in the dataframe

### Lemmatization

```python
df_selected['lemmatized'] = df_selected['nlp_doc'].apply(lambda doc: [token.lemma_ for token in doc])
df_selected
```

* Performed Lemmatization and saved it to a new column

### Names Entity Recognization

```python
df_selected['entities'] = df_selected['nlp_doc'].apply(lambda doc: [(ent.text, ent.label_) for ent in doc.ents])
df_selected
```

* Performed NER and saved it to a new column entities in the dataframe

```python
displacy.render(df_selected['nlp_doc'].iloc[0], style='ent', jupyter=True)
```

### Parsing

```python
df_selected['parsing'] = df_selected['nlp_doc'].apply(lambda doc: [(token.text, token.dep_, token.head.text) for token in doc])
df_selected
```

* Performed Parsing and saved it to a new column parsing

## Code Cell 4 (10%) – Find the top 20 frequently used nouns in 1 star reviews and 5 star reviews, respectively.

```python
noun = {}

for i in df_selected[df_selected['stars'] == 1]['pos']:
    length = len(i)
    for j in range(length):
        if i[j][1] == "NOUN":
            if i[j][0] in noun:
                noun[i[j][0]] += 1
            else:
                noun[i[j][0]] = 1

top_20 = sorted(noun.items(), key=lambda x: x[1], reverse=True)[:20]
print("Top 20 nouns for 1 Star Restaurants:")
for i,j in top_20:
    print(f"{i}: {j}")
```

```python
noun = {}

for i in df_selected[df_selected['stars'] == 5]['pos']:
    length = len(i)
    for j in range(length):
        if i[j][1] == "NOUN":
            if i[j][0] in noun:
                noun[i[j][0]] += 1
            else:
                noun[i[j][0]] = 1

top_20 = sorted(noun.items(), key=lambda x: x[1], reverse=True)[:20]
print("Top 20 nouns for 5 Star Restaurants:")
for i,j in top_20:
    print(f"{i}: {j}")
```

* Saved all the count of the noun words of how many times they have been used in the reviews in dictionary.
* Printed out Top 20 nouns for both 1 star and 5 star reviews.

## Code Cell 7 (10%) – Find the top 20 frequently used named entities from the selected reviews.


```python
all_entities = [ent for sublist in df_selected['entities'] for ent in sublist]

entity_counter = Counter([ent[0] for ent in all_entities])

top_20_entities = entity_counter.most_common(20)

print("Top 20 named entities:")
for entity, count in top_20_entities:
    print(f"{entity}: {count}")
```

* Printed out Top 20 named entities using the counter function from collection library
